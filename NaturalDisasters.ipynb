## Methods
### Data Collection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pylab as pylab
```

Importing data from the CSV File

```python
disaster = pd.read_csv("1970-2021data.csv", error_bad_lines=False)
disaster
```

```python
disaster = disaster.drop(["Dis No", "Seq", "Glide", "Disaster Group", "Disaster Subtype", "Disaster Subsubtype", "Event Name", "ISO", "Origin", "Location", "Associated Dis", "Associated Dis2", "OFDA Response", "Appeal", "Declaration", "Aid Contribution", "Dis Mag Value", "Latitude", "Longitude", "Local Time", "River Basin", "Start Day", "End Day", "No Injured", "No Affected", "No Homeless", "Total Affected", "Reconstruction Costs ('000 US$)", "Insured Damages ('000 US$)", "Total Damages ('000 US$)", "CPI", "Adm Level", "Admin1 Code", "Admin2 Code", "Geo Locations"], axis=1)
```

```python
disaster = disaster.dropna(axis=0)
```

### Analyses

In this area I explore the relationship between the Disasters and the types they have occured throughout the 50 year period, the amount of natural disasters that take part in each country, the amount of disasters per region, Countries vs the total amount of deaths, and last but not least the disaster subgroup vs the amount of deaths per subgroup.

### Distributions
Firstly I will start by telling my data to use only the columns that I want, and I will be doing a comparison between the types of Natural Disasters vs the amount of times they occured within the 50 year period that we have chosen to do.

```python
disaster.columns = ["Year", "Disaster Subgroup", "Disaster Type", "Country", "Region", "Continent", "Dis Mag Scale", "Start Year", "Start Month", "End Year", "End Month", "Total Deaths"]
plt.title("Types of Disasters ranging from 1970-2021")
plt.xticks(rotation=90, ha='right')
fig=sns.countplot(x= "Disaster Type", data=disaster)
fig = fig.set(title = "Types of Disasters ranging from 1970-2021", xlabel = 'Disaster Type', ylabel = 'Occurences')

```

### Amount of Natural Disasters vs Countries

For this next data below, I compared the amount of natural disasters within the countries that have experienced them, this will help us understand which country had it worst over the time period. Based on the graph below we can see that China and Urugauy really had some trouble over the last 50 years, facing disasters over and over again, there are a few more countries that also had it as rough; including Bangladesh, India, and the Phillipines. There was definitely a lot of south asian descent.

```python

sns.set(rc = {'figure.figsize':(50,8)})
plt.title("Amount of Natural Disasters within countries")
plt.xticks(rotation=90, ha='right')
fig = sns.countplot(x= "Country", data=disaster)
fig = fig.set(title = "Amount of Natural Disasters within countries", xlabel = 'Country', ylabel = 'Disasters')


```

### Disaster Subgroup vs Region

Here we have a group by operation of our Disaster Subgroup and the regions that they have occured in. I chose to do a plot bar as it shows the data very clear and concise. We can see that Southern Asia typically has more occurences over the 50 year period, and I assume that is because of India, Bangladesh, and China

```python
grouped = disaster["Disaster Subgroup"].groupby([disaster["Region"]])
grouped.count().plot.bar()
plt.title('Amount of Disasters per region ')
```

An analysis of this table shows that we have Southern Asia as the most affected by Disasters, it is shown that we have a well over 1400 disasters in that region. On the other end of the spectrum we have either Macedonia or the Russian Federation with between 0-50 disasters that have taken place within that 50 year range. I assume them to have meteorological damages and Hydrological damages due to being close to the Ganges Delta, and having many rivers.


### Deaths vs Country

This data below, shows us the big comparison of the amount of deaths per country, it is definitely one of the most upsetting graphs to really look at. We can see that the graph maxes out at about 300,000 deaths as an approximate. It looks like Bangladesh had one of the most casualties out of all the other countries.

```python
sns.set(rc = {'figure.figsize':(45,15)})
plt.xticks(rotation=90, ha='right')
plt.title("Deaths vs Country")
plt.ylabel("# of Deaths")
plt.xlabel("Countries")
sns.boxplot(x= "Country", y="Total Deaths", data=disaster)
```

### Disaster Subgroup vs Total Deaths

The next and final graph shows the comparison of each Subgroup vs the Total death per subgroup, we then take the average out of that, which is 310.06 and we also look for the max amount of deaths which is 300,000.

```python
plt.title("Disaster Subgroup vs Total Deaths")
sns.set(rc = {'figure.figsize':(30,30)})
plt.xticks(rotation = 0, ha='right')
fig = sns.countplot(x= "Disaster Subgroup", data=disaster)
fig = fig.set(title = "Disaster Subgroup vs Total Deaths", xlabel = 'Disaster Subgroup', ylabel = 'Total Deaths')
ds = disaster["Total Deaths"].describe()
print(f"Mean Total Deaths: {round(ds['mean'],2)}")
print(f"Max Total Deaths: {round(ds['max'],2)}")
```

Based off our last graph above we can see that Hydrological had occured the most within that 50 years period, we can see that Climatogical was actually less. Climatogical disasters started occuring more closer to 2020 than closer to 1970, we had more quality of water depreciate in Hydrological and floods occuring in meterological and we continue to do so even today.
